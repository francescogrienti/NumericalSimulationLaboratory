```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
data1 = np.loadtxt("results_1.dat") #Exponential
data2 = np.loadtxt("results_2.dat") #Lorentzian
data3 = np.loadtxt("results_3.dat") #Uniform
```

**FIG.1 - STANDARD DICE**

```python
fig = plt.figure(figsize=(12,8))

#Fit
mu, sigma = scipy.stats.norm.fit(data3[:,3])
N=scipy.stats.norm(loc=mu,scale=sigma) 
xmin=np.min(data3[:,3])
xmax=np.max(data3[:,3])
x = np.linspace(xmin,xmax,101)

ax_1 = fig.add_subplot(2, 2, 1)
ax_1.hist(data3[:,0], bins=100, density=True)
ax_1.set_title("N=1")
 
ax_2 = fig.add_subplot(2, 2, 2)
ax_2.hist(data3[:,1], bins=100, density=True)
ax_2.set_title("N=2")

ax_3 = fig.add_subplot(2, 2, 3)
ax_3.hist(data3[:,2], bins=100, density=True)
ax_3.set_title("N=10")

ax_4 = fig.add_subplot(2, 2, 4)
ax_4.hist(data3[:,3], bins=100, density=True, label="Data")
ax_4.plot(x, N.pdf(x), label="Gaussian fit", color="red")
ax_4.legend()
ax_4.set_title("N=100")
```

**FIG.2 - EXPONENTIAL DICE**

```python
fig = plt.figure(figsize=(12,8))

#Fit
mu, sigma = scipy.stats.norm.fit(data1[:,3])
N=scipy.stats.norm(loc=mu,scale=sigma) 
xmin=np.min(data1[:,3])
xmax=np.max(data1[:,3])
x = np.linspace(xmin,xmax,101)

ax_1 = fig.add_subplot(2, 2, 1)
ax_1.hist(data1[:,0], bins=100, density=True)
ax_1.set_title("N=1")
 
ax_2 = fig.add_subplot(2, 2, 2)
ax_2.hist(data1[:,1], bins=100, density=True)
ax_2.set_title("N=2")

ax_3 = fig.add_subplot(2, 2, 3)
ax_3.hist(data1[:,2], bins=100, density=True)
ax_3.set_title("N=10")

ax_4 = fig.add_subplot(2, 2, 4)
ax_4.hist(data1[:,3], bins=100, density=True, label="Data")
ax_4.plot(x, N.pdf(x), label="Gaussian fit", color="green")
ax_4.legend()
ax_4.set_title("N=100")
```
