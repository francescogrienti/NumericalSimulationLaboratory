```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
data1 = np.loadtxt("results_1.dat")
data2 = np.loadtxt("results_2.dat")
data3 = np.loadtxt("results_3.dat")
```

```python
#Parameters
M = 100000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N) 
x *= L 
mu = 0.5
var = 1/12
```

**1-Mean**

```python
#Mean
plt.figure(figsize=(12,6))
plt.errorbar(x,data1[:,0]-mu,yerr=data1[:,1])
plt.plot(x,data1[:,0]-mu, color="red")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,data1[:,0]-mu, color="red")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

**2-Standard deviation**

```python
#Standard deviation
plt.figure(figsize=(12,6))
plt.figure(figsize=(12,6))
plt.errorbar(x,data2[:,0]-var,yerr=data2[:,1])
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

**3-$\chi^2$**

```python
y = np.arange(N)+1
plt.figure(figsize=(12,6))
plt.scatter(y,data3, color="red")
plt.xlabel('#evaluations')
plt.ylabel('$\chi^2$')
plt.grid(True)
plt.show()
```

```python
k1, loc, x_1 = scipy.stats.chi2.fit(data3, floc=0)
```

```python
plt.rcParams.update({'font.size': 18})
xmin=np.min(data3)
xmax=np.max(data3)
x = np.linspace(xmin,xmax,101)
#Histogram
plt.figure(figsize=(12,6))
plt.hist(data3,bins=10, density=True, label='Data')
#Fit
W1=scipy.stats.chi2(k1,loc=0, scale=x_1) 
plt.plot(x, W1.pdf(x), label="$\chi^2$ fit", color="red")
plt.xlabel('X')
plt.ylabel('p(X)')
plt.legend()
```
