```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data1 = np.loadtxt("results_1.dat")
data2 = np.loadtxt("results_2.dat")
```

```python
M = 100000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)
x = np.arange(N) 
x *= L 
plt.figure(figsize=(12,6))
plt.errorbar(x,data1[:,0]-0.5,yerr=data1[:,1])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()

plt.figure(figsize=(12,6))
plt.errorbar(x,data2[:,0]-1/12,yerr=data2[:,1])
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```
