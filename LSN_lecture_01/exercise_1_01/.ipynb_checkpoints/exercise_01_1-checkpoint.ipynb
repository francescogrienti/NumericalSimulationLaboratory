```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data1 = np.loadtxt("results_1.dat")
data2 = np.loadtxt("results_2.dat")
data3 = np.loadtxt("results_3.dat")
```

```python
#Parameters
M = 100000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N) 
x *= L 
mu = 0.5
var = 1/12
```

**1-Mean**

```python
#Mean
plt.figure(figsize=(12,6))
plt.errorbar(x,data1[:,0]-mu,yerr=data1[:,1])
plt.plot(x,data1[:,0]-mu, color="red")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
plt.plot(x,data1[:,0]-mu, color="red")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

**2-Standard deviation**

```python
#Standard deviation
plt.figure(figsize=(12,6))
plt.errorbar(x,data2[:,0]-var,yerr=data2[:,1])
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

**3-$\chi^2$**

```python
y = np.arange(N)+1
plt.scatter(y,data3, color="red")
plt.xlabel('#evaluations')
plt.ylabel('$\chi^2$')
plt.grid(True)
plt.show()
```
