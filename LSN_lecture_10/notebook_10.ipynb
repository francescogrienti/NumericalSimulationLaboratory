This exercise was aimed at parallelizing the code for the Genetic Algorithm (GA) implemented in exercise 9. In particular, the code was parallelized in order to perform an independent GA search for each node used and exchanging every $N_{migr}$ generations the best individuals among different nodes. The MPI (Message Passing Interface) libraries were exploited for this task in the SPMD framework, meaning that the same program was run on different datasets (generated by the Random Number Generator fed with different seeds, check the code for further details and the method "initialize"). The processes on different nodes are uniquely identified by the rank, an integer $\in [0, size-1]$ in which the size is the total number of processes run. The following methods from the MPI libraries were used: 

- MPI_Init() :  method for initializing the MPI libraries; 
- MPI_Comm_size() : method for checking how many processes are active; 
- MPI_Comm_rank() : method for getting information about the rank of the process; 
- MPI_Send() : method for sending the best individual; 
- MPI_Recv() : method for receiving the best individual; 
- MPI_Finalize() : method for finalizing the MPI libraries;

The code, once parallelized, was then applied to the TSP for the italian provinces in two different scenarios: the first one with no migrations of the best individuals while in the second one the code was implemented allowing for the exchange of the best individuals.  
For both cases, $N_{population} = 1500$ and $N_{generations} = 1000$, $P_{crossover} = 0.9$ and $P_{mutations} = 0.07$. The parameter responsible for the selection operator was kept at $p = 3$. The $N_{migrations}$ was set at 3 and the code was run on 4 cores. In the following of the notebook the results are plotted.
