```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
data1 = np.loadtxt("results_radius.dat")
data2 = np.loadtxt("results_xyz.dat")
```

```python
print(data2[:,0].mean())
```

```python
M = 1000000 #Number of throws
N = 200 #Number of blocks
L = (int)(M/N) #Number of throws per block
x = np.arange(N) 
x *= L 
r = 1.5
```

```python
plt.figure(figsize=(12,6))
plt.hist(data2[:,0], bins=30, density=True)
plt.grid(True)
plt.show()
```

```python
#Mean
plt.figure(figsize=(12,6))
plt.errorbar(x,data1[:,0]-r,yerr=data1[:,1])
plt.axhline(y = 0, color = 'r', linestyle = '-')
plt.xlabel('#throws')
plt.ylabel('<r>-3/2')
plt.grid(True)
plt.show()
```

```python
X=data2[:,1]
Y=data2[:,2]
Z=data2[:,3]
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
ax.autoscale(enable=True, axis='both', tight=True)
ax.set_title('Stato fondamentale',fontweight='bold',fontsize=16)

plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.scatter(data2[:,2],data2[:,3])
plt.xlabel('#throws')
plt.ylabel('<r>-3/2')
plt.grid(True)
plt.show()
```
