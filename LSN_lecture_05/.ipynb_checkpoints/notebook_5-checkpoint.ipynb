In quantum physics a **wave function**, $\Psi$, is a mathematical description of the state of a quantum system. The wave function is a complex-valued probability amplitude, and the probabilities for the possible results of measurements made on the system can be derived from it.

For now on, let's consider the simple case of a non-relativistic single particle, without spin, in three spatial dimensions. The state of such a particle is completely described by its wave function, $\Psi(\vec{r},t)$, where $\vec{r}$ is position and $t$ is time.

For one spinless particle, if the wave function is interpreted as a probability amplitude, the square modulus of the wave function, $|\Psi(\vec{r},t)|^2$, is interpreted as the probability density that the particle is at $\vec{r}$ at time $t$. Therefore, given a PDF, it is possible to sample it using the Metropolis Algorithm, namely $M(RT)^{2}$. 

#### Hydrogen atom
The wave functions of the eigenstates of an electron in a Hydrogen atom (this is the only atom for which the Schroedinger equation has been solved exactly) are expressed in terms of spherical harmonics and generalized Laguerre polynomials. It is convenient to use spherical coordinates, and the wave function can be separated into functions of each coordinate:

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm is the Bohr radius, $L_{n-l-1}^{2l+1}$ are the generalized Laguerre polynomials of degree $n-l-1$, $n=1,2,...$ is the principal quantum number, $l=0,1, ..., n-1$ the azimuthal quantum number, $m=-l, -l+1, ..., l-1, l$ the magnetic quantum number.

For example, the ground state wave function is:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
whereas one of the three $2p$ excited state is:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
<span style="color:blue">Expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
</span>

#### Metropolis Algorithm
The Metropolis Algorithm is a Monte Carlo technique for PDF sampling. Its strength resides in the fact that it is able to sample any probability density function. It is based on a Markovian process and the following steps summarize how it was implemented in the code: 

- Choose a starting point $(x,y,z)$ in the 3D space. In this specific case, it is reasonable to start near the region where one expects to sample the majority of values;
- Random step from the starting point $(x,y,z)$. In particular, in this exercise uniform and gaussian steps were investigated and at this stage it is important to choose a $\delta$ of the step in order to satisfy the condition reported in the last step below (50% rule); 
- Accept or reject the step proposed. In order to do this, one computes the acceptance $A(\vec{x}'|\vec{x})=\min\bigg[1,\frac{p(\vec{x}')}{p(\vec{x})}\bigg]$ where $p(\vec{x})=|\Psi(\vec{x})|^2$. If a random number sampled from [0,1) is smaller than the acceptance then the new starting point of the algorithm becomes the point reached from the step and the process continues. In this way, the algorithm explores the regions of space at higher probability;
- Choose a $\delta$ value (which is, statistically speaking, the width of the distribution one is considering) in order to get, on average, 50% for the step acceptance.

For this exercise, the following values of $\delta$ were chosen:

- Gaussian transition probability: $\delta_{Ground State} = 0.75$, $\delta_{Excited State} = 1.8$;
- Uniform transition probability: $\delta_{Ground State} = 1.2$, $\delta_{Excited State} = 2.9$;

For what concerns the starting point:

- Ground state : $\vec x = (1, 1, 1)$;
- Excited state: $\vec x = (3, 3, 3)$;

**EQUILIBRATION PHASE**

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy

R_GS, x_GS, y_GS, z_GS= np.loadtxt("results_eq_GS.dat", unpack=True, usecols=(0,1,2,3))
R_ES, x_ES, y_ES, z_ES = np.loadtxt("results_eq_ES.dat", unpack=True, usecols=(0,1,2,3))

M = 1000
x = np.arange(M)
r_GS = 1.5
r_ES = 5.

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(20,6))
ax0.plot(x, R_GS[:1000], label=r'$\vec{x}_{0}=(40,40,40)$')
ax0.axhline(y=r_GS, color='r', ls='--', label='Target radius')
ax0.set_title('Ground State')
ax0.set_xlabel('#steps')
ax0.set_ylabel(r'$r$')
ax0.legend()
ax0.grid(True)

ax1.axis('off')
ax1 = fig.add_subplot(122, projection='3d')
ax1.scatter(x_GS, y_GS, z_GS, c=z_GS, marker='.')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.view_init(10, 30)
plt.show()

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(20,6))
ax0.plot(x, R_ES[:1000], label=r'$\vec{x}_{0}=(60,60,60)$')
ax0.axhline(y=r_ES, color='r', ls='--', label='Target radius')
ax0.set_title('Excited state')
ax0.set_xlabel('#steps')
ax0.set_ylabel(r'$r$')
ax0.legend()
ax0.grid(True)

ax1.axis('off')
ax1 = fig.add_subplot(122, projection='3d')
ax1.scatter(x_ES, y_ES, z_ES, c=z_ES, marker='.')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.view_init(10, 30)
plt.show()
```

The equilibration plots above report a sampling of the PDFs starting very far from the origin. They highlight how the equilibration is needed before start measuring: in fact, starting very far from the origin, the systems equilibrate after 400 steps and $\sim$ 230 steps respectively for the ground state and for the excited state. On the right hand side of the plots it can also be seen the characteristic shapes (spherical and two-lobes) of the 1s and 2p orbitals. However, as mentioned above, the simulation was run starting close to the point we expected to be the majority of samplings.  
In the simulations run in the following of the notebook, $N_{blocks} = 200$, $M = 10^{6}$ and therefore $ L = \frac{M}{N} = 5000$.
