The variational principle in Quantum Mechanics states that the expectation value of the Hamiltonian over a (good) trial quantum state has the ground state energy $E_0$ as lower bound:

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

During this exercise you will variationally optimize the ground state of a single quantum particle in a one dimensional (1D) space confined by the following external potential:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

This 1D quantum model is not analytically solvable and should be attacked by numerical methods. In order to obtain an approximate wave function for the ground state, the Variational Monte Carlo method uses a trial wave function, parametrized by a set of variational parameters. An approximate model for the ground state wave function is the following trial wave function:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

$\Psi_T^{\sigma,\mu}(x)$ depends on <span style="color:red">two variational parameters: $\sigma$ and $\mu$</span>.

### Exercise 08.1

Write a Variational Monte Carlo code for a single quantum particle in 1D which exploits the Metropolis algorithm to sample the square modulus of our trial wave function $|\Psi_T^{\sigma,\mu}(x)|^2$ using an uniform transition probability $T(x_{new}|x_{old})$. By using data blocking, the code should be able to compute the expectation value for the Hamiltonian

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

which can be used to find the parameters that minimize this quantity. Use $\hbar = 1$ and $m=1$.
<span style="color:red">Note that to measure $\langle {\hat H} \rangle_T$ you need the analytical calculation of:</span>
    
$$
\frac{{-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}} \Psi_T^{\sigma,\mu}(x)}{\Psi_T^{\sigma,\mu}(x)}
$$

In this exercise a variational Monte Carlo code was implemented using the Metropolis Algorithm to sample the trial wave function $|\Psi_T^{\sigma,\mu}(x)|^2$. In particular, for the first exercise, the variational parameters (after some trials) were fixed at $\mu = 0.83$ and $\sigma = 0.65$ with $\delta_{Metropolis} = 2.3$ for the $50%$ rule acceptance in order to get as close as possible to the functional form of $\Psi$ obtained from numerical computation (see below in the notebook for further details). 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data_acceptance = np.loadtxt("/home/francesco/NumericalSimulationLaboratory/LSN_lecture_08/exercise_8_01/results_x_hamiltonian_GS.dat")
data_energy = np.loadtxt("/home/francesco/NumericalSimulationLaboratory/LSN_lecture_08/exercise_8_01/results_energy_GS.dat")
```

```python
#Parameters
M = 1000000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N) 
```

```python
print(data_acceptance[:,0].mean())
```

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2
    
def psi2(x,mu,sigma):
    alpha = ((x-mu)/sigma)**2
    beta = ((x+mu)/sigma)**2
    N = 2*sigma*np.sqrt(np.pi)*(1+np.exp(-mu**2/sigma**2))
    return (np.exp(-alpha/2.) + np.exp(-beta/2.))**2/N

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
mu = 0.83
sigma = 0.65

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

# Plot a few things
plt.figure(figsize=(14,8))
scale = 0.3
plt.plot(x, (psi[0])**2, label="Numerical (Matricial) solution")
plt.plot(x, psi2(x, mu, sigma), label="Guessed solution")
plt.hist(data_acceptance[:,1], bins='auto', density=True, alpha=scale, label='Sampled configurations')
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.ylabel("$|\Psi_T(x)|^2$")
plt.xlim((-3,3))
plt.ylim((-0.2,0.6))
plt.grid(True)
plt.legend()
```

```python
#Parameters
M = 1000000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N)

plt.figure(figsize=(12,6))
plt.plot(x, data_energy[:,0], linestyle= 'dotted', color="green")
plt.xlabel("#blocks")
plt.ylabel("Ground state Energy")
plt.title(r'GS energy with fixed parameters $\mu = 0.83$ and $\sigma = 0.65$')
plt.grid(True)
plt.show()
```

In this exercise it was asked to extend the Variational Monte Carlo code implemented in exercise 8.1 with a Simulated Annealing algorithm in order to find the parameters $\mu$ and $\sigma$ that minimize $\langle {\hat H} \rangle_T$. The Simulated Annealing method is a thermodynamic-based technique that exploits the slowly cooling down of a system in order to end up in a global minimum state energy for the system. In fact, if the system is rapidly cooled down it could end up in a metastable state which is not a gloabl minimum for the energy landscape.  
The idea behind SA algorithm is to sample from the Boltzmann distribution $$ p(\vec{x}) = \frac{e^{-\beta L(\vec{x})}}{Z}$$ in which $L(\vec{x})$, the cost function, plays the role of the energy of the system. In particular, the algorithm moves the system to a new configuration with probability 1 if the energy of the new configuration is lower than the previous one, otherwise if the energy of the new configuration is higher then the move is accepted with probability $$e^{-\beta (L(\vec{x'}) - L(\vec{x}))}.$$  
At each step $\beta$ changes and in the limit $\beta \to +\infty$ the system finds itself in the global minimum of the energy landscape.
