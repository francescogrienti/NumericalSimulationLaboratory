```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
data_call_direct = np.loadtxt("results_call_direct.dat")
data_call_discretized = np.loadtxt("results_call_discretized.dat")
data_put_direct = np.loadtxt("results_put_direct.dat")
data_put_discretized = np.loadtxt("results_put_discretized.dat")
```

```python
M = 100000 #Number of throws
N = 200 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N) 
x *= L 
```

```python
plt.figure(figsize=(12,6))
plt.errorbar(x,data_call_direct[:,0],yerr=data_call_direct[:,1])
plt.axhline(y = 14.975790778311286, color = 'r', linestyle = '-')
plt.xlabel('$N_{steps}$')
plt.ylabel('Call price - Direct')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.errorbar(x,data_call_discretized[:,0],yerr=data_call_discretized[:,1])
plt.axhline(y = 14.975790778311286, color = 'r', linestyle = '-')
plt.xlabel('$N_{steps}$')
plt.ylabel('Call price - Discretized')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.errorbar(x,data_put_direct[:,0],yerr=data_put_direct[:,1])
plt.axhline(y = 5.4595325819072364, color = 'r', linestyle = '-')
plt.xlabel('$N_{steps}$')
plt.ylabel('Put price - Direct')
plt.grid(True)
plt.show()
```
