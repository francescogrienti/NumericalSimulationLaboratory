```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
data1 = np.loadtxt("results_1.dat")
data2 = np.loadtxt("results_2.dat")
```

```python
M = 100000 #Number of throws
N = 100 #Number of blocks
L = (int)(M/N)#Number of throws per block
x = np.arange(N) 
x *= L 
```

```python
plt.figure(figsize=(12,6))
plt.errorbar(x,data1[:,0],yerr=data1[:,1])
plt.axhline(y = 14.975790778311286, color = 'r', linestyle = '-')
plt.xlabel('$N_{steps}$')
plt.ylabel('Call price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.errorbar(x,data2[:,0],yerr=data2[:,1])
plt.axhline(y = 5.4595325819072364, color = 'r', linestyle = '-')
plt.xlabel('$N_{steps}$')
plt.ylabel('Put price')
plt.grid(True)
plt.show()
```
